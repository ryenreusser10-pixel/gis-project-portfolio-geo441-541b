## **Part 3: Reflection**

*In a brief paragraph, compare the two maps you created. Describe the visual and data model differences between the DEM datasets visualized in the topology map and the zonal statistics calculated and appended to the parcel dataset in the second map.*

The Oxford Topography map shows the elevation of Oxford. The higher elevation is noted by the color red and lower elevation in blue. The Oxford Zonal Statistics map shows the slope of each parcel. Higher slopes are in darker colors while the lower slopes are in lighter colors. The maps are great for analyzing Oxford. For example, I noticed that the areas with higher elevation are also the areas lowest in slope and vice versa. This makes sense because Oxford is mostly a plateau/flat. In the areas where Oxford is low in elevation and high in slope are areas that have more hills. Together, these maps display the landscape of Oxford being relatively flat and highly elevated while the outer edges and parks are at lower elevation with a higher slope. The location is perfect for a college campus since the ground is flat with low slope, great place to build large buildings and sidewalks.
